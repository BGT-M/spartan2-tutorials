```python
import spartan as st
```

```python
import scipy.sparse as ssp
```

```python
data = st.loadTensor(path='./inputData/soc-Epinions1.tensor.gz', header=None, col_idx=[0, 1], col_types=[int, int], sep='\t')
mapper = st.DenseIntMapper()
tensor = data.toSTensor(hasvalue=False, mappers={0: mapper, 1: mapper})
N = mapper._idx
tensor.shape = (N, N)
```

```python
print(tensor)
```

## Run as a model 

```python
summa = st.Summarize(tensor)
```

```python
res = summa.run(T=5)
```

## Run as a task

```python
model = st.Summarization.create(tensor, st.SumPolicy.Summarize, 'summarize')
model.run(T=5)
```
